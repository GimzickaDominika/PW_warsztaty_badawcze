## Together AI

```python
from together import Together
```

```python
with open("homework_1_ang.txt", 'r') as file:
    lines = file.readlines()
```

```python
client = Together(api_key="-")
```

```python
prompt = f"In a moment i will sent an encrypted text in English. There will 10 senteces in total. Each sentence will be encrypted with Caesar method, but the shift is unknown. Sentence to decrypt: {lines[0]}"
    
response = client.chat.completions.create(
    model="meta-llama/Llama-3.3-70B-Instruct-Turbo",
    messages=[{"role": "user", "content": prompt}],
)
print(response.choices[0].message.content)
```

```python
for i, line in enumerate(lines):
    print(f"__________Sentence {i+1}__________")
    
    prompt = f"In a moment i will sent an encrypted text in English. The method is unknown. Sentence to decrypt: {line}"

    response = client.chat.completions.create(
        model="deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=8000,
    )
    print(response.choices[0].message.content)
```
