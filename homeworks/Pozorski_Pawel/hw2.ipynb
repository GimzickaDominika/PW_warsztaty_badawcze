```python
import numpy as np

def cronbach_alpha(items):
    """
    Oblicza współczynnik Alfa Cronbacha dla podanych danych.

    :param items: Lista list (macierz odpowiedzi), gdzie każda kolumna to jedna zmienna.
    :return: Wartość Alfa Cronbacha.
    """
    items = np.array(items)
    N = items.shape[1]
    
    variances = np.var(items, axis=0, ddof=1)
    total_variance = np.var(np.sum(items, axis=1), ddof=1)
    
    alpha = (N / (N - 1)) * (1 - np.sum(variances) / total_variance)
    return alpha
```
