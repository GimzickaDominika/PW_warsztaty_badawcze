Zobaczmy kod (polska wersja)

```python
with open('../homework_1.txt', 'r') as file:
    tasks = [line.strip() for line in file.readlines()]

print('\n'.join(f"{i+1:4.0f}. {task}" for i, task in enumerate(tasks)))
```

```python
def get_last_promt(task: str) -> dict:
    return {
        "role": "user",
        "content": f"Oto zaszyfrowany ciąg: `{task}`. Tylko wynik w postaci tekstu po rozszyfrowaniu.",
    }
```

Zdefiniujmy sobie prompty. Ze względu na ograniczone zasoby dla wszystkich zapytan zostawiłem prompt który eksperymentalnie na podstawie gpt3.5 turbo działał mi najlepiej.

```python
prompts = [
    {
        "name": "podstawowy",
        "history": [
            {
                "role": "system",
                "content": "Jesteś doświadczonym kryptografem i lingwistą specjalizującym się w łamaniu szyfrów. Analizujesz teksty pod kątem statystycznym, gramatycznym i kontekstowym, aby znaleźć wzorce i możliwe metody deszyfrowania.",
            },
            {
                "role": "user",
                "content": "Mam tekst w języku polskim zaszyfrowany prostym szyfrem (np. szyfr Cezara). Wiem że wynik również jest w języku polskim, oraz że ma tyle samo słów co zaszyfrowant tekst oraz że każde z odszyfrowanych słów jest tej samej długości.",
            },
        ],
    }
]
```

```python
for task in tasks:
    task_history = get_last_promt(task)

    for prompt in prompts:
        print("\n".join(h["content"] for h in prompt["history"] + [task_history]))

        print()
```

```python
results = []

for task in tasks:
    task_history = get_last_promt(task)

    for prompt in prompts:
        completion = client.chat.completions.create(
            model=model,
            messages=prompt["history"] + [task_history],
        )

        results.append(
            {
                "prompt": prompt,
                "task": task,
                "result": completion.choices[0].message.content,
                "completion": completion,
            }
        )
```

```python
import json
from copy import deepcopy

results_to_save = deepcopy(results)
for result in results_to_save:
    result.pop("completion")
with open('together.json', 'w', encoding='utf-8') as f:
    json.dump(results_to_save, f)
```
