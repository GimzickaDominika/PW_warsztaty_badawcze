```python
# LLM postawiony lokalnie - model Deepseek-R1 (7b (4.7 GB)) został wcześniej pobrany za pomocą Ollama 
```

```python
import ollama
# importuje ollame

zdania_path = "zdania.txt"
# Wczytanie danych z poprawnym kodowaniem
with open(zdania_path, "r", encoding="utf-8") as f:
    zdania = f.readlines()

zdania = [z.strip() for z in zdania]
```

```python
# Chciałem sprawdzić czy podanie dodatkowych informacji / kontekstu wpłynie znacząco na wynik
```

```python
stream = ollama.chat(
    model="deepseek-r1",
    messages=[{"role": "user", "content": f"decrypt the encoded message: '{zdania[0]}'"}],
    stream=True,
    )

for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
```

```python
# Koniec outputu promptu: 
"""decrypt the encoded message: '{zdanie do odszyfrowania}'"""

# Output:
"""
Given this confusion, but sticking to the first two words as correctly decrypting, I think the answer intended is:

The decrypted message would have 'The' and 'placebo' as the first two words. The third might not make sense or could be incorrect due to a typo or mistake in the encryption.
</think>

The decrypted encoded text is "The placebo efffect."""
```
