1. Wstęp teoretyczny (wikipedia):

Szyfr Cezara – jedna z najprostszych technik szyfrowania. Jest to rodzaj szyfru podstawieniowego, w którym każda litera tekstu jawnego (niezaszyfrowanego) zastępowana jest inną, oddaloną od niej o stałą liczbę pozycji w alfabecie, literą (szyfr monoalfabetyczny), przy czym kierunek zamiany musi być zachowany. Nie rozróżnia się przy tym liter dużych i małych. 

Matematycznie: x - pozycja przed szyfrowaniem, n - shift, y - pozycja w zakodowanym
- Kodowanie: y = x + n mod (liczba znaków w alfabecie)
- Dekodowanie: x = y - n mod(liczba znaków w alfabecie)
